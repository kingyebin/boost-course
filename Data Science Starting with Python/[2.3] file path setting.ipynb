```python
# 모든 csv 파일 불러오기
glob('data/소상공인시장진흥공단_상가(상권)정보_20240630/*.csv')
```

```python
# "소상"으로 시작하는 모든 파일 불러오기
glob('data/소상공인시장진흥공단_상가(상권)정보_20240630/소상*.csv')
```

```python
# "_"가 포함되는 모든 파일 불러오기
glob('data/소상공인시장진흥공단_상가(상권)정보_20240630/*_*.csv')
```

```python
# 모든 파일 리스트 형태로 불러오기
file_csv = glob('data/소상공인시장진흥공단_상가(상권)정보_20240630/*.csv')
file_csv
```

```python
file_list = []
for file_csv_name in file_csv:
    print(file_csv_name)
    temp = pd.read_csv(file_csv_name)
    file_list.append(temp)
```

```python
# 전체 파일 출력
file_list
```

```python
# 첫번째 DataFrame 로드
file_list[0]
```

```python
# 마지막 DataFrame 로드
file_list[-1]
```

```python
# column 명이 같다면 하나의 DataFrame으로 합칠 수 있다.
df = pd.concat(file_list)
df
```
